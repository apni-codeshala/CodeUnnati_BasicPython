### Plot Subplots in Matplotlib

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1)  # Parameters 1st -> 2 rows, 2nd -> 1 column, 3rd -> 1st position
plt.plot(x, y)

# PLot - 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2)    # Parameters 1st -> 2 rows, 2nd -> 1 column, 3rd -> 2st position
plt.plot(x, y)
```

### Pie Chart

```python
from matplotlib import pyplot as plt
import numpy as np

# Creating Dataset

cars = ['Audi', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating PLot

fig = plt.figure(figsize = (10, 7))
plt.pie(data, labels = cars)

plt.show()
```

### Histogram and Binnig

```python
import matplotlib.pyplot as plt
import numpy as np

x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]

plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')

plt.tick_params(axis='x', labelsize=18, labelrotation=-60, labelcolor='turquoise')
plt.tick_params(axis='y', labelsize=12, labelrotation=20, labelcolor='orange')

plt.show()
```

## Some problems on numpy

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])
print(data)

## Mean
print("Mean :",np.mean(data))

## Median
print("Median :",np.median(data))

print("Standard Deviation :",np.std(data))

print("Max :",np.max(data))

print("Min :",np.min(data))
```

```python
import numpy as np

data = np.array([15, 25, 35, 45, 55])
print(data)

mean = np.mean(data)
standard_deviation = np.median(data)
print("Mean :", mean)
print("Standard Deviation :", standard_deviation)

print("Values greater than mean :",data[data > mean])
```

```python
import mpl_toolkits as mplot3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = plt.axes(projection = '3d')

z = np.linspace(0, 1, 1000)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')

plt.show()
```
