```python
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

## Scatter Plot

```python
sns.scatterplot(tips['total_bill'], tips['tip'])
```

```python
sns.lineplot(tips['total_bill'], tips['tip'])
```

```python
sns.regplot(tips['total_bill'], tips['tip'])
```

```python
 sns.barplot(tips['total_bill'], tips['tip'])
```
